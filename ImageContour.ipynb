```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2

# Contour로 탐지시 마커도 contour로 탐지됨 
# 곡선만 따로 탐지하는 방법이 필요할거같음
def contour():
  img = cv2.imread('/content/drive/MyDrive/CrackDetection/target.jpg')
  imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  ret, thr = cv2.threshold(imgray, 127,255,0)
  tmp = cv2.findContours(thr, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

  contours = tmp[0] if len(tmp) == 2 else tmp[1]
  print(len(contours))

  cnt = contours[4]
  area = cv2.contourArea(cnt)
  perimeter = cv2.arcLength(cnt, True)

  cv2.drawContours(img, [cnt], -1, (255,255,0), 3)
  print('면적 : ', area)
  print('길이 :', perimeter)
    
  cv2_imshow(img)

  cv2.waitKey(0)
  cv2.destroyAllWindows()

contour()
```

```python
import numpy as np
import cv2

def moment():
  img = cv2.imread('/content/drive/MyDrive/CrackDetection/test2.png')
  imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  ret, thr = cv2.threshold(imgray, 127,255,0)
  tmp = cv2.findContours(thr, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  contours = tmp[0] if len(tmp) == 2 else tmp[1]

  contour = contours[0]
  mmt = cv2.moments(contour)

  for key, val in mmt.items():
    print('%s:\t%.5f' %(key, val))

  cx = int(mmt['m10']/mmt['m00'])
  cy = int(mmt['m01']/mmt['m00'])

  print(cx,cy)

moment()
```
